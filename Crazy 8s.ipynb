```python
import random 

cards = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
suits = ['Clubs', 'Diamonds', 'Hearts', 'Spades']
deck = []
pile = []

# create a deck with 13 cards for each of the 4 suits
for i in range(len(suits)): 
    for j in range(len(cards)):
        deck.append( (cards[j], suits[i]) ) 
    
random.shuffle(deck)

# deal 5 cards to each player
player1_hand = deck[0:5]
del deck[0:5]

player2_hand = deck[5:10]
del deck[5:10]

print(player1_hand)
print(player2_hand)

starter_card = deck[0]

# make sure the first card flipped over isn't an 8
while starter_card in [('8', 'Clubs'), ('8', 'Diamonds'), ('8', 'Hearts'), ('8', 'Spades')]:
    random.shuffle(deck)
    starter_card = deck[0]
    
# add the starter card to the new pile
pile.append(starter_card)
del deck[0]
current_card = starter_card
```

```python
print('the current card is:', current_card)

# play the game until either player doesn't have any cards in their hand
while len(player1_hand) > 0 or len(player2_hand) > 0:
        
    # assume player 1 goes first
    valid_move_possible = False
    while not valid_move_possible:

        print('player 1 hand is:', player1_hand)

        # ask player 1 for their card choice 
        number = input('what number card does player 1 want to play? ')
        suit = input('what suit? ')

        card_choice = number, suit

        # have player 1 draw if they do not have a valid move 
        if not( card_choice[1] == current_card[1] or card_choice[0] == current_card[0] or card_choice[0] == '8'):
            player1_hand.append(deck[0]) 
            del deck[0]
        else:
            valid_move_possible = True

    # add the card to the pile once player 1 has made a valid move  
    index = player1_hand.index(card_choice)
    del player1_hand[index]
    pile.append(card_choice)
    
    # make the current card = to last one in the pile
    current_card = pile[-1]  
    print('the current card is:', current_card)
    
    # assume player 2 goes next 
    valid_move_possible = False
    while not valid_move_possible:

        print('player 2 hand is:', player2_hand)

        # ask player 2 for their card choice 
        number = input('what number card does player 2 want to play? ')
        suit = input('what suit? ')

        card_choice = number, suit

        # have player 2 draw if they do not have a valid move 
        if not( card_choice[1] == current_card[1] or card_choice[0] == current_card[0] or card_choice[0] == '8'):
            player2_hand.append(deck[0]) 
            del deck[0]
        else:
            valid_move_possible = True

    # add the card to the pile once player 2 has made a valid move  
    index = player2_hand.index(card_choice)
    del player2_hand[index]
    pile.append(card_choice)
    
    # make the current card = to last one in the pile
    current_card = pile[-1] 
    print('the current card is:', current_card)
    
    # the player says 'Crazy 8' if they only have one card left in their hand
    if len(player1_hand) == 1: 
        print('player 1 says Crazy 8!')
    elif len(player2_hand) == 1: 
        print('player 2 says Crazy 8!')
```

```python
# check which player has no cards left and say they are the winner
if len(player1_hand == 0):
    print('player 1 won!')
elif len(player2_hand == 0):
    print('player 2 won!')
```
